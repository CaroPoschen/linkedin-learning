```python
MonthlySales = []
```

```python
with open('MonthlySales.csv', 'r') as f:
    reader = csv.DictReader(f)
    for row in reader:
        MonthlySales.append(row)
```

```python
for a in MonthlySales:
    print(a)
```

```python
#print keys
for a in MonthlySales:
    print(a.keys())
```

```python
#print keys and values
for a in MonthlySales:
    for key, value in a.items():
        print(key + ": ", value)
    print('\n')
```
